# Web Scraping - Python
## Using BeautifulSoup

- FIY: All webpages are made up of HTML & CSS. (I believe many know this)
- Using Inspect, we can see all the elements of a webpage.

- I have taken Flipkart page for example. This can be mostly done on any site.

### Load in Necessary Libraries

```python
import requests #This is to load the webpages
from bs4 import BeautifulSoup as bs 
#If any of the libraries didn't get imported, then pip install requests (or) pip install beautifulsoup4
```

### Load a Webpage

```python
#Load the webpage content
r = requests.get('https://www.flipkart.com/audio-video/headphones/pr?sid=0pm%2Cfcn&otracker=categorytree&p%5B%5D=facets.connectivity%255B%255D%3DBluetooth&fm=neo%2Fmerchandising&iid=M_9eac98fb-c6d3-47a7-86b8-fb80eddf0324_2_372UD5BXDFYS_MC.R08R6GB1Q1BI&otracker=hp_rich_navigation_2_2.navigationCard.RICH_NAVIGATION_Electronics~Audio~Bluetooth%2BHeadphones_R08R6GB1Q1BI&otracker1=hp_rich_navigation_PINNED_neo%2Fmerchandising_NA_NAV_EXPANDABLE_navigationCard_cc_2_L2_view-all&cid=R08R6GB1Q1BI')
#Convert to a beautifulsoup object
soup = bs(r.content)

#print out the HTML content
#print(soup) #This would be hard to read or even to look at
print(soup.prettify()) #This might give a better understanding
```

```python
type(soup)
```

```python
selection_class = 's1Q9rs'
titles = soup.find_all('a', {'class': selection_class}) #Find the class of the text/info you are looking for.
```

```python
len(titles)
```

```python
titles[0]['title'] #get title that is within the tag
```

```python
products =[]

for item in titles:
    products.append(item['title'])
print(products[:5])
```

```python
len(products)
```

```python
titles[0]['href'] #get link of the product thats in the tag
```

```python
product_url = 'https://www.flipkart.com'+titles[0]['href']
print(product_url)
```

```python
type(titles[0]['href'])
```

```python
titles[0]['href'].rpartition('?')[0] #This is to take everything before the question mark '?', 
#since that part is enough for the link, all other are just navigation and tracking
```

```python
products_url =[]

for item in titles:
    products_url.append(('https://www.flipkart.com'+item['href'].rpartition("?")[0]))
print(products_url[:5])
```

```python
images = soup.find_all('img', {'class': '_396cs4 _3exPp9'})
len(images)
```

```python
images[0]
```

```python
images[0]['src'] #Get the image source of the products
```

```python
imagel = []
for img in images:
    imagel.append(img['src'])
    
print(imagel[:5])
```

```python
rating = '_3LWZlK'
ratings = soup.find_all('div', {'class': rating})
len(ratings)
```

```python
ratings[0].text
```

```python
ratingsl = []
for r in ratings:
    ratingsl.append(r.text)
    
print(ratingsl[:5])
```

```python
rating_v = '_2_R_DZ'
rating_volumes = soup.find_all('span', {'class': rating_v})
len(rating_volumes)
```

```python
rating_volumes[0].text
```

```python
ratingsvol = []
for n in rating_volumes:
    ratingsvol.append(n.text.strip('()'))
    
ratingsvol[:5]
```

```python
price = '_30jeq3'
prices = soup.find_all('div', {'class': price})
len(prices)
```

```python
prices[0].text
```

```python
pricelist =[]
for x in prices:
    pricelist.append(x.text[1:]) #This is to remove the Rupee symbol, such that conversion to int will be easier
pricelist[:5]
```

```python
color_type = '_3Djpdu'
color_types = soup.find_all('div',{'class':color_type})
len(color_types)
```

```python
color_types[0].text
```

```python
colorntype=[]
for k in color_types:
    colorntype.append(k.text)
colorntype[:5]
```

### Save data and Export as CSV

```python
import pandas as pd
```

```python
dct = {'Product':products, 'Product_Link':products_url, 'Product_Image':imagel, 
       'Rating':ratingsl, 'Rating_Volumes':ratingsvol, 'Price':pricelist, 'Color_Type':colorntype}
```

```python
df = pd.DataFrame(dct)
df.head()
```

```python
df.to_csv('Headphones_Buylist.csv', index=None)
```
